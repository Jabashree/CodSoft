```python
import tkinter as tk
import random

# Define the questions, options, and correct answers
questions = [
    {
        "question": "What is the capital of France?",
        "options": ["Madrid", "Rome", "Berlin", "Paris"], 
        "correct_option": "Paris",
    },
    {
        "question": "What is 2 + 2?",
        "options": ["3", "4", "5", "6"],
        "correct_option": "4",
    },
    {
        "question": "Who wrote 'Romeo and Juliet'?",
        "options": ["Charles Dickens", "Jane Austen", "William Shakespeare", "Leo Tolstoy"],
        "correct_option": "William Shakespeare",
    },
]

# Create a function to display a random question
def show_random_question():
    global current_question
    current_question = random.choice(questions)
    question_label.config(text=current_question["question"])
    for i in range(4):
        option_buttons[i].config(text=current_question["options"][i])

# Create a function to check the answer
def check_answer(option):
    if option == current_question["correct_option"]:
        result_label.config(text="Correct!")
    else:
        result_label.config(text="Incorrect!")

# Create the main window
root = tk.Tk()
root.title("Random Quiz Generator")

# Create and configure widgets
question_label = tk.Label(root, text="", font=("Arial", 12))
result_label = tk.Label(root, text="", font=("Arial", 12))

option_buttons = []
for i in range(4):
    option_buttons.append(tk.Button(root, text="", command=lambda i=i: check_answer(current_question["options"][i])))

next_question_button = tk.Button(root, text="Next Question", command=show_random_question)

# Pack widgets
question_label.pack()
for i in range(4):
    option_buttons[i].pack()
result_label.pack()
next_question_button.pack()

# Show the first question
show_random_question()

# Start the tkinter main loop
root.mainloop()
```
